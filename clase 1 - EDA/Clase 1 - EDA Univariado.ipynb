Los datos pueden ser representados como una matriz $n\times d$, con $n$ filas y $d$ columnas, en este caso mencionaremos que la matriz tiene $n$ instancias o puntos, y posee $d$ atributos. <br>
Denotaremos $x_i$ (x minuscula) como la fila o instancia i, mientras que $X_k$ (X mayuscula) denotara la columna o atributo k. <br>
El tamaño de la matriz se refiere al numero de instancias, mientras que la cantidad de atributos se refiere a la dimensionalidad de los datos.

# Tipos de atributos

Los atributos se categorizaran según el dominio de sus datos, existen 2 categorias principales:

1. **Numericos** <br>
    ***
    1. **Numericos de intervalo**: Representan unidades ordenadas donde conocemos las diferencias exactas entre los valores. El problema con los datos de valores de intervalo es que podemos sumar y restar, pero no podemos multiplicar, dividir o calcular razones. Debido a que no existe un cero verdadero. Por ejemplo la temperatura, tiene sentido mencionar que si ayer estabamos a 10°C, y hoy a 20°C, entonces hubo un aumento de 10°C, mas seria erroneo decir que hace el doble de calor.
    2. **Numericos de relación**: También son unidades ordenadas que tienen la misma diferencia. Pero ademas existe un cero absoluto, con lo que operaciones como la multiplicacion y division tienen sentido. Por ejemplo si una persona tiene 10 años y otra 20 años, tiene sentido decir que una persona tiene el doble de edad que la otra.
2. **Categoricos** 
    ***
    1. **Categoricos nominales**: Son datos que no poseen orden, por lo que solo operaciones de igualdad tienen sentido. Por ejemplo los colores, podemos decir que un objeto tiene el mismo color o si tiene uno diferente, mas no que un color es mayor o menor al otro.
    2. **Categoricos ordinales**: Sos datos que poseen orden, mas no hay diferencias entre estos. Por lo que comparaciones de igualdad y desigualdad (mayor que, menor que) tienen sentido. Un ejemplo es la educación, ya que se puede ordenar de menor grado a mayor grado, mas no tiene sentido intentar restar o ayar una diferencia cuantitativa entre doctorado y pregrado.

# Analisis univariado

## Atributo numerico

Para el analisis univariado de atributos numericos, se encuentran varias metricas descriptivas que parten de la visualización estadistica de los datos, como media, moda, mediana, varianza, desviación estandar, cuartiles, etc. Ademas de otras metricas probabilisticas como distribuciones de densidad o acumulativas.

```python
uni_example = np.array(iris_x["sepal length (cm)"])
print(uni_example)
print("La media es:", uni_example.mean())
print("La mediana es:", np.median(uni_example))
print("La moda es:", iris_x["sepal length (cm)"].mode()[0])
```

```python
print(np.array(iris).shape)
np.array(uni_example).shape
```

```python
def probabilidad_acumulativa(n, data):
    for index, dat in enumerate(np.sort(data)):
        if dat > n:
            return index/data.shape[0]
def probabilidad_densidad(n, data):
    sum = 0
    for dat in data:
        if dat == n:
            sum += 1
    return sum/data.shape[0]

print("Probabilidad de que un dato sea menor o igual a 5.9:", probabilidad_acumulativa(5.9, uni_example))
print(np.where(uni_example<5.9)[0].shape[0]/uni_example.shape[0])
print("Probabilidad de que un dato sea 5:", probabilidad_densidad(5.0, uni_example))
print(np.where(uni_example==5)[0].shape[0]/uni_example.shape[0])
```
